```python
def prediction_process(X,y,stop_index):
    # 训练回归模型
    X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=19500, shuffle=False)
    
    model_knn = KNeighborsRegressor()  # 建立贝叶斯岭回归模型对象

    model_names = ['KNN']  # 不同模型的名称列表
    model_dic = [model_knn]  # 不同回归模型对象的集合

    pre_y_list = []  # 各个回归模型预测的y值列表
    for model in model_dic:  # 读出每个回归模型对象
        pre_y_list.append(model.fit(X_train, y_train).predict(X_test[stop_index]))  # 将回归训练中得到的预测y存入列表，要分训练集和预测集
    # 模型效果指标评估

    
    return pre_y_list, y_test[stop_index]
```

```python
def DCE_predictor(iteration, re_Long, stop_index):
    for i in range(iteration):
        prediction_step = i+1
        X_Long = np.array(re_Long.head(min_len-prediction_step))
        y_Long = np.array(re_Long.iloc[:,-1].tail(min_len-prediction_step))
        y_Long = y_Long.reshape([min_len-prediction_step,1])

        pre_y_list, y_test = prediction_process(X_Long,y_Long,stop_index)

        test = abs(y_test-pre_y_list[0])

    #     y_no_noise_loc = np.where(test < 0.1)[0]
    #     y_test_filtered = y_test[y_no_noise_loc].reshape(-1,1)
    #     y_pred_filtered = pre_y_list[0][y_no_noise_loc]

        if i ==0:
            temp_test = pd.DataFrame(y_test)
            temp_pred = pd.DataFrame(pre_y_list[0])
        else:
            temp_test = pd.concat([temp_test,pd.DataFrame(y_test)], ignore_index=True, axis=1)
            temp_pred = pd.concat([temp_pred,pd.DataFrame(pre_y_list[0])], ignore_index=True, axis=1)
    
    return temp_test,temp_pred
```
